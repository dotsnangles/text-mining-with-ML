```python
import pandas as pd
```

```python
fileNames = [
'감성.csv',
'개그.csv',
'드라마.csv',
'무협사극.csv',
'순정.csv',
'스릴러.csv',
'스포츠.csv',
'액션.csv',
'일상.csv',
'판타지.csv',
]

dfs= []
```

```python
# pd.read_csv(fileNames[8], index_col=[0])
# pd.read_csv(fileNames[9])

df1 = pd.read_csv(fileNames[0])
df2 = pd.read_csv(fileNames[1])
df3 = pd.read_csv(fileNames[2])
df4 = pd.read_csv(fileNames[3])
df5 = pd.read_csv(fileNames[4], index_col=[0])
df6 = pd.read_csv(fileNames[5])
df7 = pd.read_csv(fileNames[6])
df8 = pd.read_csv(fileNames[7], index_col=[0])
df9 = pd.read_csv(fileNames[8])
df10 = pd.read_csv(fileNames[9]).drop(['Unnamed: 0', '제목.1', '장르.1', '회차.1'], axis=1)
df10.insert(0, '회차', df10.pop('회차'))
df10.insert(0, '제목', df10.pop('제목'))
df10.insert(0, '장르', df10.pop('장르'))
```

```python
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10]
new_df = pd.concat(dfs)
```

```python
new_df.set_index(pd.Series(range(len(new_df))), inplace=True)
```

```python
new_df.to_csv('네이버웹툰장르통합.csv', index=False)
```
