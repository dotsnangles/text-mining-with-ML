```python
# 3글자 장르

def addCols(fileName):
  df = pd.read_csv(fileName, index_col=[0])
  df.insert(0, '장르', fileName[-7:-4])
  df.insert(1, '제목', fileName[:-8])
  df.insert(2, '회차', [i for i in range(1, len(df)+1)])
  return df
```

```python
fileNames = [
  'MLB카툰_스포츠.csv',
  '가비지타임_스포츠.csv',
  '격기3반_스포츠.csv',
  '그린보이_스포츠.csv',
  '빅맨_스포츠.csv',
  '빌드업_스포츠.csv',
  '윈드브레이커_스포츠.csv',
  '자율공상축구탐구만화_스포츠.csv',
  '탑코너_스포츠.csv',
]

dfs= []

for fileName in fileNames:
  dfs.append(addCols(fileName))
```

```python
new_df = pd.concat(dfs)
```

```python
new_df.set_index(pd.Series(range(len(new_df))), inplace=True)
```

```python
new_df.to_csv('스포츠.csv', index=False)
```
